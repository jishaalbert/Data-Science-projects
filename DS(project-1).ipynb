# age group most of the employees belong to.

# which team and position, spending in terms of salary is high.

```python
spending=employees_df.groupby(['Team','Position'])['Salary'].sum()
```

```python
high=spending.idxmax()
print(high)
```

# correlation between age and salary , represent it visually.

```python
correlation=df['Age'].corr(df['Salary'])
print("Correlation =",correlation)
```

```python
plt.scatter(df['Age'],df['Salary'],color="red")
plt.title('Age vs Salary')
plt.xlabel('Age')
plt.ylabel('Salary')
```
